```python
import os, string, sys, glob
import numpy as np
import pyaml, json
import logging
logger = logging.getLogger()
from IPython.display import HTML, display

import imageio as io

import rfinder as rfinder
import rfinder_plots as rfi_pl
rfiPL = rfi_pl.rfi_plots()

#animations in notebook
from matplotlib import rc


print('\t+------+\n\t  Done \n\t+------+')
```

### Load and modify parameter file

```python
print('\t+------+\n\t RFInder \n\t Parameter File \n\t+------+')

rfi_par = rfinder.rfinder()

print('\t+------+\n\t Parameters Loaded \n\t+------+')

#print parameter file 
#print(json.dumps(rfi_par.cfg_par, indent=4, sort_keys=True))
```

```python
# Set working directory & filenames
#rfi_par.cfg_par['general']['workdir'] = '/home/maccagni/RFI/'

# Enable / disable modules
rfi_par.cfg_par['rfi']['enable'] = True
rfi_par.cfg_par['plots']['enable'] = True

#MSFILE
rfi_par.cfg_par['general']['msname'] = ['fa1_conc.ms']
rfi_par.cfg_par['general']['field'] = 0

rfi_par.set_dirs()

# CHUNKS
#create also tables of data binned in frequency and time

#plotting options
rfi_par.cfg_par['plots']['altaz_gif'] = True

print('\t+------+\n\t Parameters modified \n\t+------+')
```

### Alt/Az gifs

```python

if rfi_par.cfg_par['rfi']['RFInder_mode']=='use_flags':
    filenames = glob.glob(rfi_par.cfg_par['general']['altazplotdir']+'/AltAZ_flags*')

    tmp_arr=[]
    for i in xrange(0,len(filenames)):
        #print filenames[i]
        #print 'aaa'
        tmp = filenames[i].split('flags')[1]
        tmp_arr.append(tmp.split('MHz')[0])
    tmp_arr.sort()
        
    filenames = [tmp[0]+'flags' + s for s in tmp_arr] 
    filenames = [s + 'MHz.png' for s in filenames] 
elif rfi_par.cfg_par['rfi']['RFInder_mode']=='rms_clip':
    filenames = glob.glob(rfi_par.cfg_par['general']['altazplotdir']+'/AltAZ_rfi*')

    tmp_arr=[]
    for i in xrange(0,len(filenames)):
        #print filenames[i]
        #print 'aaa'
        tmp = filenames[i].split('_rfi')
        tmp_arr.append(tmp[1].split('MHz')[0])
    tmp_arr.sort()
        
    filenames = [tmp[0]+'_rfi' + s for s in tmp_arr] 
    filenames = [s + 'MHz.png' for s in filenames] 


out_animation = rfi_par.cfg_par['general']['moviedir']+'AltAz_movie.gif'
images=[]

for filename in filenames:
    images.append(io.imread(filename))
    
io.mimsave(out_animation,images,duration=1.5)
    
print('\t+------+\n\t Done \n\t+------+')
```

### Time chunks 2D gifs

```python
#select files

if rfi_par.cfg_par['rfi']['RFInder_mode']=='use_flags':
    filenames = glob.glob(rfi_par.cfg_par['general']['timeplotdir2D']+'/flags_base*')
elif rfi_par.cfg_par['rfi']['RFInder_mode']=='rms_clip':
    filenames = glob.glob(rfi_par.cfg_par['general']['timeplotdir2D']+'/rfi_base*')

tmp_arr=[]
for i in xrange(0,len(filenames)):
        tmp = filenames[i].split('base_')[1]
        tmp = filenames[i].split('base_')[1]
        tmp_arr.append(tmp.split('m.png')[0])
tmp_arr.sort()

tmp = filenames[0].split('base_')
filenames = [tmp[0]+'base_' + s for s in tmp_arr] 
filenames = [s + 'm.png' for s in filenames] 
out_animation = rfi_par.cfg_par['general']['moviedir']+'Time_2Dplot_movie.gif'
    
images=[]
for filename in filenames:
    images.append(io.imread(filename))
io.mimsave(out_animation,images,duration=1.5)

print('\t+------+\n\t Done \n\t+------+')
```

### Time chunks 1D gifs

```python
#select files
filenames_flags = glob.glob(rfi_par.cfg_par['general']['timeplotdir1D']+'/flags_*')

tmp_arr=[]

for i in xrange(0,len(filenames_flags)):
    if rfi_par.cfg_par['rfi']['RFInder_mode']=='use_flags':
        if len(filenames_flags[i].split('sl_rfi.'))>1:
            continue
    elif rfi_par.cfg_par['rfi']['RFInder_mode']=='rms_clip':
        if  len(filenames_flags[i].split('sl_flags.'))>1:
            continue
    tmp = filenames_flags[i].split('flags_')[1]
    tmp_arr.append(tmp.split('m_sl_rfi.png')[0])

tmp_arr.sort()
tmp = filenames_flags[0].split('flags_')

filenames_flags = [tmp[0]+'flags_' + s for s in tmp_arr]
if rfi_par.cfg_par['rfi']['RFInder_mode']=='use_flags':
    filenames_flags = [s + 'm_sl_flags.png' for s in filenames_flags] 
elif rfi_par.cfg_par['rfi']['RFInder_mode']=='rms_clip':
    filenames_flags = [s + 'm_sl_rfi.png' for s in filenames_flags] 

out_animation = rfi_par.cfg_par['general']['moviedir']+'TimeChunks_1D_flags.gif'
    
    
images=[]
for filename in filenames_flags:
    images.append(io.imread(filename))
io.mimsave(out_animation,images,duration=1.5)

#select files
filenames_noise = glob.glob(rfi_par.cfg_par['general']['timeplotdir1D']+'/noise_*')
tmp_arr=[]
for i in xrange(0,len(filenames_noise)):
    if rfi_par.cfg_par['rfi']['RFInder_mode']=='use_flags':
        if len(filenames_noise[i].split('sl_rfi.'))>1:
            continue
    elif rfi_par.cfg_par['rfi']['RFInder_mode']=='rms_clip':
        if  len(filenames_noise[i].split('sl_flags.'))>1:
            continue
    tmp = filenames_noise[i].split('noise_')[1]
    tmp_arr.append(tmp.split('m_sl_rfi.png')[0])

tmp_arr.sort()
tmp = filenames_noise[0].split('noise_')

filenames_noise = [tmp[0]+'noise_' + s for s in tmp_arr]
if rfi_par.cfg_par['rfi']['RFInder_mode']=='use_flags':
    filenames_noise = [s + 'm_sl_flags.png' for s in filenames_noise] 
elif rfi_par.cfg_par['rfi']['RFInder_mode']=='rms_clip':
    filenames_noise = [s + 'm_sl_rfi.png' for s in filenames_noise] 


out_animation = rfi_par.cfg_par['general']['moviedir']+'TimeChunks_1D_noise.gif'

images=[]
for filename in filenames_noise:
    images.append(io.imread(filename))
io.mimsave(out_animation,images,duration=1.5)

filenames_noisefactor = glob.glob(rfi_par.cfg_par['general']['timeplotdir1D']+'/noisefactor_*')
#select files
tmp_arr=[]
for i in xrange(0,len(filenames_noisefactor)):
    if rfi_par.cfg_par['rfi']['RFInder_mode']=='use_flags':
        if len(filenames_noisefactor[i].split('sl_rfi.'))>1:
            continue
    elif rfi_par.cfg_par['rfi']['RFInder_mode']=='rms_clip':
        if  len(filenames_noisefactor[i].split('sl_flags.'))>1:
            continue
    tmp = filenames_noisefactor[i].split('noisefactor_')[1]
    tmp_arr.append(tmp.split('m_sl_rfi.png')[0])

tmp_arr.sort()
tmp = filenames_noisefactor[0].split('noisefactor_')

filenames_noisefactor = [tmp[0]+'noisefactor_' + s for s in tmp_arr]
if rfi_par.cfg_par['rfi']['RFInder_mode']=='use_flags':
    filenames_noisefactor = [s + 'm_sl_flags.png' for s in filenames_noisefactor] 
elif rfi_par.cfg_par['rfi']['RFInder_mode']=='rms_clip':
    filenames_noisefactor = [s + 'm_sl_rfi.png' for s in filenames_noisefactor] 

out_animation = rfi_par.cfg_par['general']['moviedir']+'TimeChunks_1Dnoisefactor.gif'
images=[]
for filename in filenames_noisefactor:
    images.append(io.imread(filename))
io.mimsave(out_animation,images,duration=1.5)



print('\t+------+\n\t Done \n\t+------+')
```
